```python
import sys
import os
import itertools
sys.path.append(os.path.abspath('../'))
from state_constrainer_YY import state_constrainer
from termcolor import colored
```

```python
m = state_constrainer('boundedStack.sol')
```

```python
metodos = ["push","pop","top","isEmpty"]
precondiciones = [x+"_precondition" for x in metodos]
def call_pre():
    for cond in precondiciones:
        m.callContractFunction(cond,tx_sender=m.witness_account)
```

```python
states = list(itertools.product([0,1],repeat=len(metodos)))
def repr_state(state):
    text = ""
    for x,method in zip(state,metodos):
        text += '_'+method if x else ""
    if text == "":
        text = "vacio"
    return text
```

```python
call_pre()
```

```python
for state in states:
    count = m.generateTestCases(keys=precondiciones,targets=state,testcaseName=f"STATE_{repr_state(state)}")
    if count > 0:
        print(colored(f"found {count} testcases that reach {repr_state(state)} initial state"),'green')
    else:
        print(colored(f"found no testcases for {repr_state(state)} initial state"),'red')
```
